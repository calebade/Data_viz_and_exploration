## Data Preprocessing and Cleaning

### TODO:
* Save the uncleaned copy of the dataframe
* Check of columns in the datasets that need cleaning like the `YearlyChange`, `WorldShare`, etc columns
* Handle Missing Data as you deem fit - [Pandas Missing Data Guide](https://pandas.pydata.org/docs/user_guide/missing_data.html)
* Correct the data types where necessary
* Create a `wrange_data` function that reads and cleans the data
* Make sure to include informative comments


```python
# Define the wrange_data function
def wrange_data(file_path):
    uncleaned_df = pd.read_csv(file_path)

    # Save a copy to be cleaned
    cleaned_df = uncleaned_df.copy()

    # remove all forms of null values
    cleaned_df.replace(['N.A.', '', 'nan'], np.nan, inplace=True)

    # Clean columns by removing '%' and converting to float
    columns_to_clean = ['YearlyChange', 'WorldShare', 'UrbanPop%']
    for col in columns_to_clean:
        cleaned_df[col] = cleaned_df[col].str.rstrip('%').astype(float) / 100

    # Handle missing data
    cleaned_df['Fert.Rate'].fillna(cleaned_df['Fert.Rate'].median(), inplace=True)
    cleaned_df['MedianAge'].fillna(cleaned_df['MedianAge'].median(), inplace=True)
    cleaned_df['UrbanPop%'].fillna(cleaned_df['UrbanPop%'].median(), inplace=True)

    # remove duplicates if any
    cleaned_df.drop_duplicates(inplace = True)

    # Return the cleaned DataFrame and the uncleaned copy
    return uncleaned_df, cleaned_df


uncleaned_data, cleaned_data = wrange_data('/content/WorldPopulation2023.csv')
cleaned_data.head()
```

```python
top_10_populated = cleaned_data.nlargest(10, 'Population2023')[['Country', 'Population2023']]
least_10_populated = cleaned_data.nsmallest(10, 'Population2023')[['Country', 'Population2023']]

# Reverse the order of data to flip the plots
top_10_populated = top_10_populated[::-1]
least_10_populated = least_10_populated[::-1]

# Create subplots for top 10 most populated and least 10 populated countries
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot for top 10 most populated countries
ax1.barh(top_10_populated['Country'], top_10_populated['Population2023'])
ax1.set_xlabel('Population (2023)')
ax1.set_ylabel('Country')
ax1.set_title('Top 10 Most Populated Countries')

# Plot for least 10 populated countries
ax2.barh(least_10_populated['Country'], least_10_populated['Population2023'], color='orange')
ax2.set_xlabel('Population (2023)')
ax2.set_ylabel('Country')
ax2.set_title('Least 10 Populated Countries')

plt.tight_layout()
plt.show()
```

# *The two plots above show the top 10 most populated and top 10 least populated countries respectively. some further insights or questions that we could ask is what other features are correlated with the population and what combination of features would indicate overpopulation or underpopulation.*

```python
# Heatmap of the correlation between numeric features
plt.figure(figsize=(10, 8))
sns.heatmap(cleaned_data.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```

# *This gives the correlation between the various columns in our data. e.g.columns like Fertility rate and yearly change are positively correlated while columns like median age and yearly change strongly correlate negatively.*

```python
df=cleaned_data
# Relationship between Migrants and Population Density
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Migrants(net)', y='Density(P/Km)', alpha=0.6)
plt.title('Migrants vs. Population Density')
plt.xlabel('Migrants (net)')
plt.ylabel('Population Density (P/Km)')
plt.show()
```

# *This plot shows that there is generally no correlation between the population density and the net migrants which can be further confirmed from the heatmap generated*

```python
# Relationship between Median Age and Yearly Change
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='MedianAge', y='YearlyChange', alpha=0.6)
plt.title('Median Age vs. Yearly Change')
plt.xlabel('Median Age')
plt.ylabel('Yearly Change')
plt.show()
```

# *Based on my observation from the heatmap, the median age and yearly change are negatively correlated by a value of -0.73, this plot gives us a visual on this correlation.*

```python
# Relationship between Fertility Rate and Yearly Change
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Fert.Rate', y='YearlyChange', alpha=0.6)
plt.title('Fertility Rate vs. Yearly Change')
plt.xlabel('Fertility Rate')
plt.ylabel('Yearly Change')
plt.show()
```

# *The Fertility Rate positively correlates with the Yearly Change depicted by the general upward trend of our scatterplot and as also seen from the heatmap by a value of 0.71*

```python

# indicators for overpopulation and underpopulation
overpopulation_threshold = 500
underpopulation_threshold = 50

cleaned_data['Overpopulation'] = cleaned_data['Density(P/Km)'] > overpopulation_threshold
cleaned_data['Underpopulation'] = cleaned_data['Density(P/Km)'] < underpopulation_threshold

# top overpopulated countries
top_overpopulated = cleaned_data[cleaned_data['Overpopulation']].nlargest(10, 'Density(P/Km)')[['Country', 'Density(P/Km)']]

# top underpopulated countries
top_underpopulated = cleaned_data[cleaned_data['Underpopulation']].nsmallest(10, 'Density(P/Km)')[['Country', 'Density(P/Km)']]

# Reverse the order of data to flip the plots
top_overpopulated = top_overpopulated[::-1]
top_underpopulated = top_underpopulated[::-1]

# subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot for top overpopulated countries
ax1.barh(top_overpopulated['Country'], top_overpopulated['Density(P/Km)'], color='red')
ax1.set_xlabel('Population Density')
ax1.set_ylabel('Country')
ax1.set_title('Top Overpopulated Countries')

# Plot for top underpopulated countries
ax2.barh(top_underpopulated['Country'], top_underpopulated['Density(P/Km)'], color='green')
ax2.set_xlabel('Population Density')
ax2.set_ylabel('Country')
ax2.set_title('Top Underpopulated Countries')

plt.tight_layout()
plt.show()
```

# *2 new columns were generated and then ploted to show us the top countries that are overpopulated and underpopulated. An interesting insight from this is that the countries with high population are not necessarily overpopulated as would have been suspected, likewise for the least populated countries and underpopulation. It is only a balance between the population and the land area*

```python
# Visualize Yearly Change vs. Population
plt.figure(figsize=(8, 6))
sns.scatterplot(data=cleaned_data, x='YearlyChange', y='Population2023', hue='Overpopulation')
plt.title('Yearly Change vs. Population')
plt.xlabel('Yearly Change')
plt.ylabel('Population (2023)')
plt.legend(title='Overpopulation')
plt.show()
```

# *There is no correlation between the Population and yearly change*

```python
# Visualize Median Age vs. Population
plt.figure(figsize=(8, 6))
sns.scatterplot(data=cleaned_data, x='MedianAge', y='Population2023', hue='Overpopulation')
plt.title('Median Age vs. Population')
plt.xlabel('Median Age')
plt.ylabel('Population (2023)')
plt.legend(title='Overpopulation')
plt.show()
```

# *There is also no correlation between the population and the median age*

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=cleaned_data, x='UrbanPop%', bins=20, kde=True)
plt.title('Urban Population Percentage Distribution')
plt.xlabel('Urban Population Percentage')
plt.ylabel('Frequency')
plt.show()
```
